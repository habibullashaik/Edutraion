```python
import pandas as pd
#first Load the dataset
data = pd.read_csv('IMDb Movies India.csv',encoding='ISO-8859-1')
#Now quick Analyze the dataset columns of top rows 
data.head()
```

```python
#we got multiple Actors like Actor1,2,and 3
# combine them into a simple column Actors
data['Actors'] = data['Actor 1'].fillna('')+data['Actor 2'].fillna('')+data['Actor 3'].fillna('')
```

```python
#Drop unwanted columns
data = data[['Genre','Director','Actors','Rating']]

#Drop Rows with missing values
data=data.dropna(subset=['Rating'])
```

```python
data.head()
```

```python
#Now i got the cleaned data to process forther
#Spliting the data into features and target

x=data[['Genre','Director','Actors']]
y=data['Rating']
```

```python
#Now time to split the data for the training and testing purpose
#for this import train_test_split method from sklearn

from sklearn.model_selection import train_test_split
#80% data for training and remaining 20% data for testing
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

```

```python
#Design the pipeline for the model
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

Categorical_Features = ['Genre','Director','Actors']
Categorical_Transformer = OneHotEncoder(handle_unknown='ignore')

#NOw Column transfer
preprocessor = ColumnTransformer(
    transformers=[('cat',Categorical_Transformer,Categorical_Features)]
)
```

```python
#Now Build and train the regression model
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression

Pipeline_model = Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('regressor',LinearRegression())
])
Pipeline_model .fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
#Prediction time
y_prediction = Pipeline_model.predict(x_test)
#Model evaluation
mean_squared_error = mean_squared_error(y_test,y_prediction)
r2_score = r2_score(y_test,y_prediction)

print(f"Mean Squared Error : {mean_squared_error:.2f}")
print(f"R2 scroe : {r2_score:.2f}")
```

```python
#New real words data to predict the rating
new_movie_Prediction = pd.DataFrame({
    'Genre':['Drama'],
    'Director' : ['Ovais Khan'],
    'Actors':['Rasika DugalVivek GhamandeArvind Jangid	']
})

predicting_rating = Pipeline_model.predict(new_movie_Prediction)

print(f"Predting the Rating for new Movie : {predicting_rating[0]}")
```
